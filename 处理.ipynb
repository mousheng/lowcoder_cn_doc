```python
import os
import re


# 定义一个函数来处理每个文件
def process_md_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
    # 使用正则表达式删除以---开始和以---结束的文本
    content = re.sub(r"\n^---.*?---\n", "", content, flags=re.MULTILINE | re.DOTALL)
    lines = content.split("\n")
    title = lines[0][2:]
    title2 = lines[1].strip() if lines[1].strip() != "" else lines[2].strip()
    index = 1 if lines[1].strip() != "" else 2
    if lines[0].startswith("# ") and title == title2:
        lines.pop(index)

    # 保存修改后的内容回到文件中
    with open(file_path, "w", encoding="utf-8") as file:
        file.write("\n".join(lines).replace("\n\n\n", "\n\n"))

# 处理文件引用
def process_md_image(file_path, deep):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
    path = '../' * deep
    # content = content.replace('''​![](assets/''', f"![]({path}assets/")
    content = content.replace('''"../../.gitbook/assets/''', f'"{path}assets/')

    # 保存修改后的内容回到文件中
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)

# 处理链接
def process_md_href(file_path, deep):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
    if deep == 0:
        path = './'
    else:
        path = '../'
    content = content.replace('https://majiang.co/docs/', f"{path}")

    # 保存修改后的内容回到文件中
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)

# 遍历文件夹下的所有.md文件
folder_path = "/home/mou/locoder_cn_doc/"  # 将此路径替换为你的文件夹路径
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith(".md"):
            subPath = root.replace(folder_path, "")
            if subPath == "":
                deep = 0
            else:
                deep = len(subPath.split("/"))
            file_path = os.path.join(root, file)
            # process_md_file(file_path)
            process_md_image(file_path, deep)
            # process_md_href(file_path, deep)
            print(f"处理文件: {file_path}")

print("完成处理所有.md文件")
```
